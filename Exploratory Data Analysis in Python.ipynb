# chapter 2 : Data Cleaning and Imputation

# Addressing missing data

```python
planes = pd.read_csv(r"E:\courses\Data Analysis\Python\Data Analysis with Python\7 . Exploratory Data Analysis in Python\planes.csv")
display(planes.head())
```

```python
print(planes.info())
```

```python
# To know which columns don't have values (NA)
print(planes.isna().sum())
```

```python
# First step to clean data : to make a threshold to know which columns is lower than 5% of the total value
# Second step to clean data : if the number of the NA values is greater than 5% of total values :
#      1. convert it to the mean,median or mode 
#      2. if the columns is sub_groups : which is the columns consists of multi categories so each category has it's mean or median so convert to mean or median according to this column like in the Airline and Price in this example
```

```python
# Take the threshold 
threshold = len(planes)*0.05
print(threshold)
```

```python
# To compare each column with the threshold and if it's lower , it will be removed
g= planes.columns[planes.isna().sum() <= threshold]
print(g)
```

```python
# To drop all columns that have value lower than the threshold and updata the dataframe with inplace =Tr
planes.dropna(subset = g ,inplace = True)
print (planes.isna().sum())
```

```python
print(planes["Additional_Info"].value_counts())
```

```python
sns.boxplot(data=planes,x="Airline",y="Price")
plt.xticks(rotation=90)
plt.show()
```

```python
# Now we will drop the Additional_Info column 
planes = planes.drop("Additional_Info",axis=1)
print (planes.isna().sum())
```

```python
# Use the the sub_group step to convert the price column accordint each category in the Airline column
airline_prices = planes.groupby("Airline")["Price"].median()
print (airline_prices)
print("\n")
prices_dict= airline_prices.to_dict()
print (prices_dict)
```

```python
# Fill all NA value in the Price column with the median in the Airline column
planes["Price"] = planes["Price"].fillna(planes["Airline"].map(prices_dict))
print (planes.isna().sum())
```

# Converting and analyzing categorical data

```python
# Filter the DataFrame for object columns
non_numeric = planes.select_dtypes("object")

# Loop through columns
for col in non_numeric.columns:
  
  # Print the number of unique values
  print(f"Number of unique values in {col} column: ", non_numeric[col].nunique())
```

```python
# Create a list of categories
flight_categories = ["Short-haul", "Medium", "Long-haul"]

# Create short-haul values
short_flights = "^0h|^1h|^2h|^3h|^4h"

# Create medium_flights
medium_flights = "^5h|^6h|^7h|^8h|^9h"

# Create long_flights
long_flights = "10h|11h|12h|13h|14h|15h|16h"
```

```python
# Create conditions for values in flight_categories to be created
conditions = [
    (planes["Duration"].str.contains(short_flights)),
    (planes["Duration"].str.contains(medium_flights)),
    (planes["Duration"].str.contains(long_flights))
]

# Apply the conditions list to the flight_categories
planes["Duration_Category"] = np.select(conditions, 
                                        flight_categories,
                                        default="Extreme duration")

# Plot the counts of each category
sns.countplot(data =planes,x="Duration_Category")
plt.show()
```

```python
# creating a new column called Price_Category to know the ticket type from the price
twenty_fifth = planes["Price"].quantile(0.25)
median = planes["Price"].median()
seventy_fifth = planes["Price"].quantile(0.75)
maximum = planes["Price"].max()
labels =["Economy","Premium Economy","Business Class","First Class"]
bins = [0,twenty_fifth,median ,seventy_fifth,maximum]
planes["Price_Category"]= pd.cut(planes["Price"],labels=labels,bins=bins)
print (planes[["Price","Price_Category"]].head())
```

```python
sns.countplot(data=planes,x="Airline",hue="Price_Category")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.histplot(data =planes,x="Price_Category",y="Price")
plt.show()
```

# Wroking with numeric data

```python
# Extract hours and minutes into separate columns
planes[['Hours', 'Minutes']] = planes['Duration'].str.extract(r'(\d+)h ?(\d*)m?')

# Convert the extracted columns to numeric values
planes['Hours'] = pd.to_numeric(planes['Hours'], errors='coerce')
planes['Minutes'] = pd.to_numeric(planes['Minutes'].replace('', 0), errors='coerce')
planes["Minutes"] = planes["Minutes"] / 60

# Combine 'Hours' and 'Minutes' into a new column 'TotalMinutes'
planes['TotalHours'] = planes['Hours']+ planes['Minutes']

display (planes.head())
```

```python
# Preview the column
print(planes["TotalHours"].head())

# Remove the string character
# planes["TotalHours"] = planes["Duration"].str.replace("h", "")

# Convert to float data type
planes["TotalHours"] = planes["TotalHours"].astype(float)

# Plot a histogram
sns.histplot(data=planes,x="TotalHours")
plt.show()
```

```python
# Price standard deviation by Airline
planes["airline_price_st_dev"] = planes.groupby("Airline")["Price"].transform(lambda x:x.std())

print(planes[["Airline", "airline_price_st_dev"]].value_counts())
```

```python
# Median Duration by Airline
planes["airline_median_duration"] = planes.groupby("Airline")["TotalHours"].transform(lambda x:x.median())

print(planes[["Airline","airline_median_duration"]].value_counts())
```

```python
# Mean Price by Destination
planes["price_destination_mean"] = planes.groupby("Destination")["Price"].transform(lambda x:x.mean())
print(planes[["Destination","price_destination_mean"]].value_counts())
```

# Handling outliers

```python
# Plot a histogram of flight prices
sns.histplot(data=planes, x="Price")
plt.show()

# Display descriptive statistics for flight duration
print(planes["TotalHours"].describe())
```

```python
# Find the 75th and 25th percentiles
price_seventy_fifth = planes["Price"].quantile(0.75)
price_twenty_fifth = planes["Price"].quantile(0.25)

# Calculate iqr : Calculate the IQR, storing it as prices_iqr.
prices_iqr = price_seventy_fifth - price_twenty_fifth

# Calculate the thresholds : Calculate the upper and lower outlier thresholds.
upper = price_seventy_fifth + (1.5 * prices_iqr)
lower = price_twenty_fifth - (1.5 * prices_iqr)

# Subset the data : Remove the outliers from planes.
planes = planes[(planes["Price"] > lower) & (planes["Price"] < upper)]

print(planes["Price"].describe())
```

# Considerations for categorical data


```python
salaries = pd.read_csv(r"E:\courses\Data Analysis\Python\Data Analysis with Python\7 . Exploratory Data Analysis in Python\ds_salaries_clean.csv")
display(salaries.head())
```

```python
# Print the relative frequency of Job_Category
print(salaries["Designation"].value_counts(normalize=True))
```

```python
# Cross-tabulate Job_Category and Company_Size
print(pd.crosstab(salaries["Designation"], salaries["Company_Size"],values = salaries["Salary_USD"],aggfunc="mean"))
```

# Generating new features

```python
# Find the 25th percentile
twenty_fifth = salaries["Salary_USD"].quantile(0.25)

# Save the median
salaries_median = salaries["Salary_USD"].median()

# Gather the 75th percentile
seventy_fifth = salaries["Salary_USD"].quantile(0.75)

print(twenty_fifth, salaries_median, seventy_fifth)
```

```python
# Create salary labels
salary_labels = ["entry", "mid", "senior", "exec"]

# Create the salary ranges list
salary_ranges = [0, twenty_fifth, salaries_median, seventy_fifth, salaries["Salary_USD"].max()]

# Create salary_level
salaries["salary_level"] = pd.cut(salaries["Salary_USD"],
                                  bins=salary_ranges,
                                  labels=salary_labels)

# Plot the count of salary levels at companies of different sizes
sns.countplot(data=salaries,x="Company_Size",hue="salary_level")
plt.show()
```

# Generating Hypotheses

```python
# Filter for employees in the US or GB
usa_and_gb = salaries[salaries["Employee_Location"].isin(["US","GB"])]

# Create a barplot of salaries by location
sns.barplot(data=usa_and_gb,x="Employee_Location",y="Salary_USD")
plt.show()
```
